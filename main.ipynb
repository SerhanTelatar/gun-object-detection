```python
import os
from collections import Counter
import cv2

size= Counter()

img_dir = "/Users/macbook/Documents/GitHub/gun-object-detection/Handgun Detection.v2i.yolov8/train/images"

for img in os.listdir(img_dir):
    img = os.path.join(img_dir, img)
    img = cv2.imread(img)

    if img is not None:
        size[img.shape] +=1
print(size)
```
